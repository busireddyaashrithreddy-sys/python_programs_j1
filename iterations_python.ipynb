```python
#write a program to print 1 to 50 numbers using for loop?
for num in range(1,51):
    print(num)
```

```python
#write a program to iterate name using for loop
name="shivaji"
for char in name:
    print(char)
```

```python
#write a program to find the even numbers in between 100 to 150 using for loop?
for n in range(100,151):
    if (n%2==0):
       print(n,end=",")
```

```python
#write a program to find the odd nubers in between 100 to 150 using for loop?
for n in range(100,151):
    if (n%2==1):
        print(n,end=",")
```

```python
#write a program to find the vowels from strinng using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if (char in "aeiouAEIOU"):
        print(char)
```
