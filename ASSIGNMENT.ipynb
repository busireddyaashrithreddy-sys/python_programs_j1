```python
 1. Positive / Negative / Zero
num = float(input("Enter a number: "))

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
 2. Even / Odd
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
 3. Greatest of two numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print(a, "is greater")
else:
    print(b, "is greater")
    
 4. Smallest of three numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a < b:
    if a < c:
        print(a, "is the smallest")
    else:
        print(c, "is the smallest")
else:
    if b < c:
        print(b, "is the smallest")
    else:
        print(c, "is the smallest")

 5. Leap year check
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

```python
====iteration====
1.write a python program to print numbers from 1 to 10 using a for loop.
2.write a python program to calculate the sum of first 20 natural numbers using a loop.
3.write a python program to print all even numbers between 1 to 50.
4.write a python program to find the factoral of a number using a loop.
5.write a python program to count the number of digits in a given number using while loop.     
```

```python
 1. Print 1 to 10
for i in range(1, 11):
    print(i)

 2. Sum of first 20 natural numbers
total = 0
for i in range(1, 21):
    total += i
print("Sum =", total)

 3. Print even numbers from 1 to 50
for i in range(1, 51):
    if i % 2 == 0:
        print(i)

 4. Factorial of a number
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial =", fact)

 5. Count digits in a number
num = int(input("Enter a number: "))
count = 0
while num != 0:
    num //= 10
    count += 1
print("Number of digits =", count)
```
