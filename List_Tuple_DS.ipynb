#list in data structures
1. list is collection of elements enclosed by []
2. list a mutable - editable
3. it's allowed duplicate elements
4. each element separated by ,

```python
# create list
names=["shiva","vikram","vikas","divya"]
age=[25,26,27,25.8,27.7]
print(names)
print(age)
print(type(names))
print(type(age))
```

```python
#comcatenation of list
a=[10,20,30,40,50]
b=[40,50,60,70,80]
result=a+b
print(result)
```

```python
#repeating list
ages=[10,20,30,35,45]
print(ages*15)
```

```python
# accessing elements from list - index
ages=[15,16,17,18,19,26,51]
print(ages)
print(ages[4])
print(ages[-3])
```

```python
#methods or functions in list
print(dir(list))
```

```python
#description
print(help(list))
```

```python
a=[10,20,30,40,50]
print(a)
a.append(100)
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.extend([100,200,300])
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.insert(1,100)
print(a)
```

```python
#remove elements from list - pop(),remove(),clear()
a=[10,20,30,40,50]
print(a)
a.pop(4)
print(a)
```

```python
#sorting techniques - sort(reverse=true or reverse=false)
a=[6,4,5,3,1,2]
print(a)
a.sort(reverse=True) # true means descending , false means acending
print(a)
```
